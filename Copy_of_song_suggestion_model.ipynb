```python
from textblob import TextBlob
import webbrowser

# Step 1: Define emoji to mood map
emoji_moods = {
    "😊": "happy",
    "😢": "sad",
    "😡": "angry",
    "😍": "romantic",
    "😌": "relaxed"
}

# Step 2: Define mood → song mapping
mood_songs = {
    "happy": {
        "songs": [
            "Happy - Pharrell Williams",
            "Can't Stop the Feeling! - Justin Timberlake",
            "Good Time - Owl City"
        ],
        "url": "https://open.spotify.com/track/60nZcImufyMA1MKQY3dcCH"
    },
    "sad": {
        "songs": [
            "Someone Like You - Adele",
            "Fix You - Coldplay",
            "Let Her Go - Passenger"
        ],
        "url": "https://open.spotify.com/track/7gOdHgIoIKoe4i9Tta6qdD"
    },
    "angry": {
        "songs": [
            "Lose Yourself - Eminem",
            "Breaking the Habit - Linkin Park",
            "Stronger - Kanye West"
        ],
        "url": "https://open.spotify.com/track/4xkOaSrkexMciUUogZKVTS"
    },
    "relaxed": {
        "songs": [
            "Weightless - Marconi Union",
            "Better Together - Jack Johnson",
            "Sunflower - Post Malone"
        ],
        "url": "https://open.spotify.com/track/3KkXRkHbMCARz0aVfEt68P"
    },
    "romantic": {
        "songs": [
            "Perfect - Ed Sheeran",
            "All of Me - John Legend",
            "Just the Way You Are - Bruno Mars"
        ],
        "url": "https://open.spotify.com/track/0tgVpDi06FyKpA1z0VMD4v"
    }
}

# Step 3: AI Sentiment Analysis from Text
def get_ai_mood(text):
    sentiment = TextBlob(text).sentiment.polarity
    if sentiment > 0.5:
        return "happy"
    elif sentiment > 0.1:
        return "relaxed"
    elif sentiment < -0.5:
        return "angry"
    elif sentiment < 0:
        return "sad"
    else:
        return "romantic"

# Step 4: Gather inputs
text_input = input("How are you feeling? (Write something about your day): ").strip()
emoji_input = input("Enter an emoji to show your feeling (😊, 😢, 😡, 😍, 😌): ").strip()

# Step 5: Process each input
text_mood = get_ai_mood(text_input)
emoji_mood = emoji_moods.get(emoji_input)

# Step 6: Choose final mood (emoji preferred, fallback to text)
final_mood = emoji_mood if emoji_mood else text_mood

# Step 7: Recommend songs and open Spotify
if final_mood in mood_songs:
    print(f"\n🎧 Your mood is detected as: {final_mood.capitalize()}")
    print("Here are some song suggestions for you:")
    for song in mood_songs[final_mood]["songs"]:
        print(f" - {song}")

    print("\n🎶 Opening Spotify for you...")
    webbrowser.open(mood_songs[final_mood]["url"])
else:
    print("\n❌ Mood not recognized. Try again with a valid emoji or expressive text.")
```

```python
import torch
import clip
from PIL import Image
import webbrowser

# Load the CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Define moods and corresponding Spotify URLs
mood_songs = {
    "happy": {
        "url": "https://open.spotify.com/track/60nZcImufyMA1MKQY3dcCH",
        "songs": ["Happy - Pharrell Williams", "Good Time - Owl City", "Can't Stop the Feeling - JT"]
    },
    "sad": {
        "url": "https://open.spotify.com/track/7gOdHgIoIKoe4i9Tta6qdD",
        "songs": ["Someone Like You - Adele", "Fix You - Coldplay", "Let Her Go - Passenger"]
    },
    "angry": {
        "url": "https://open.spotify.com/track/4xkOaSrkexMciUUogZKVTS",
        "songs": ["Lose Yourself - Eminem", "Breaking the Habit - Linkin Park", "Stronger - Kanye West"]
    },
    "relaxed": {
        "url": "https://open.spotify.com/track/3KkXRkHbMCARz0aVfEt68P",
        "songs": ["Sunflower - Post Malone", "Weightless - Marconi Union", "Better Together - Jack Johnson"]
    },
    "romantic": {
        "url": "https://open.spotify.com/track/0tgVpDi06FyKpA1z0VMD4v",
        "songs": ["Perfect - Ed Sheeran", "All of Me - John Legend", "Just the Way You Are - Bruno Mars"]
    }
}

# CLIP mood labels
mood_labels = list(mood_songs.keys())

# Get user input
text_input = input("📝 How are you feeling? (Describe in a sentence): ").strip()
emoji_input = input("🙂 Now type your emoji (😊, 😢, 😡, 😍, 😌): ").strip()

# Combine both into one prompt for CLIP
combined_prompt = f"{text_input} {emoji_input}"

# Tokenize mood labels
text_tokens = clip.tokenize(mood_labels).to(device)

# Encode input
with torch.no_grad():
    user_token = clip.tokenize([combined_prompt]).to(device)
    user_features = model.encode_text(user_token)
    mood_features = model.encode_text(text_tokens)

    # Normalize
    user_features /= user_features.norm(dim=-1, keepdim=True)
    mood_features /= mood_features.norm(dim=-1, keepdim=True)

    # Compute similarity
    similarities = (user_features @ mood_features.T).squeeze(0)
    best_idx = similarities.argmax().item()
    detected_mood = mood_labels[best_idx]

# Output result
print(f"\n🤖 AI-detected mood: {detected_mood.capitalize()}")
print("🎧 Suggested songs:")
for song in mood_songs[detected_mood]["songs"]:
    print(f" - {song}")

print("\n🎶 Opening Spotify...")
webbrowser.open(mood_songs[detected_mood]["url"])
```

```python
from langdetect import detect
from googletrans import Translator
from textblob import TextBlob

translator = Translator()

# Step 1: User input
text = input("🔤 Enter your text (any language): ")

# Step 2: Detect language
source_lang = detect(text)
print(f"🌍 Detected Language Code: {source_lang}")

# Step 3: Ask for target language
print("\n💡 Example target codes: en = English, hi = Hindi, fr = French, te = Telugu, mn=malayalam")
target_lang = input("➡ Enter target language code: ")

# Step 4: Translate to target language
translated_target = translator.translate(text, src=source_lang, dest=target_lang)
print(f"\n🌐 Translated Text ({target_lang}): {translated_target.text}")

# Step 5: Translate to English for mood analysis
translated_english = translator.translate(text, src=source_lang, dest='en')
english_text = translated_english.text
print(f"\n📝 (Internal Use) English Text for Sentiment Analysis: {english_text}")

# Step 6: Analyze mood using TextBlob
blob = TextBlob(english_text)
polarity = blob.sentiment.polarity  # -1 to 1
print(f"📈 Polarity Score: {polarity}")

# Step 7: Suggest song
if polarity > 0.3:
    mood = "Happy"
    song = "🎵 'Happy' by Pharrell Williams"
elif polarity < -0.3:
    mood = "Sad"
    song = "🎵 'Fix You' by Coldplay"
else:
    mood = "Neutral/Calm"
    song = "🎵 'Let It Be' by The Beatles"

print(f"\n😊 Mood Detected: {mood}")
print(f"🎧 Recommended Song: {song}")
```
